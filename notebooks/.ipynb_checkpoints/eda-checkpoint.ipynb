```python
import os
import json
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from PIL import Image

# Paths
PROCESSED_DATA_DIR = "D:/AIDI/Object Detection Project/data/processed"
TRAIN_DIR = os.path.join(PROCESSED_DATA_DIR, "train")
VAL_DIR = os.path.join(PROCESSED_DATA_DIR, "val")
TEST_DIR = os.path.join(PROCESSED_DATA_DIR, "test")
```

```python
def load_annotations(dir_path):
    with open(os.path.join(dir_path, "annotations.json"), 'r') as f:
        data = json.load(f)
    return data["images"], data["annotations"]

train_images, train_annotations = load_annotations(TRAIN_DIR)
val_images, val_annotations = load_annotations(VAL_DIR)
test_images, test_annotations = load_annotations(TEST_DIR)
```

```python
print(f"Train Images: {len(train_images)}")
print(f"Validation Images: {len(val_images)}")
print(f"Test Images: {len(test_images)}")
```

```python
def get_category_distribution(annotations):
    category_count = Counter([ann['category_id'] for ann in annotations])
    return category_count

train_category_dist = get_category_distribution(train_annotations)
print("Training Category Distribution:", train_category_dist)
```

```python
plt.bar(train_category_dist.keys(), train_category_dist.values())
plt.xlabel("Category ID")
plt.ylabel("Frequency")
plt.title("Category Distribution in Training Data")
plt.show()
```

```python
def visualize_image_with_annotations(image_data, annotations, images_dir):
    image_id = image_data['id']
    image_path = os.path.join(images_dir, image_data['file_name'])
    
    # Open the image
    img = Image.open(image_path)
    plt.imshow(img)
    plt.axis('off')
    
    # Plot bounding boxes
    for ann in annotations:
        if ann['image_id'] == image_id:
            bbox = ann['bbox']  # [x, y, width, height]
            x, y, w, h = bbox
            plt.gca().add_patch(plt.Rectangle((x, y), w, h, edgecolor='red', facecolor='none', linewidth=2))
    
    plt.show()

# Example: Visualize a random training image
import random
random_image = random.choice(train_images)
visualize_image_with_annotations(random_image, train_annotations, os.path.join(TRAIN_DIR, "images"))
```

```python
bbox_sizes = [ann['bbox'][2] * ann['bbox'][3] for ann in train_annotations]  # Width * Height
plt.hist(bbox_sizes, bins=50)
plt.xlabel("Bounding Box Area")
plt.ylabel("Frequency")
plt.title("Bounding Box Size Distribution")
plt.show()
```
